```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fileName = "CY35-05_1-#3"
df = pd.read_csv(f"csv_data/Dataset_1_NCA_battery/{fileName}.csv")
```

```python
df_cycle_2 = df
plt.plot(df_cycle_2['time/s'], df_cycle_2['Ecell/V'], label='Ecell/V')
plt.xlabel('Time (s)')
plt.ylabel('Ecell (V)')
plt.title('Ecell vs Time')
plt.legend()
plt.show()
```
